```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("./DailyDelhiClimateTest.csv")
df = df[['date', 'meantemp']]
df.head()
```

```python
plt.scatter(df['date'], df['meantemp'], color='green')
```

```python
xl = df['meantemp'].ewm(alpha=0.5).mean()
```

```python
xl
```

```python
xl2 = df['meantemp'].ewm(alpha=0.9).mean()
xl3 = df['meantemp'].ewm(alpha=0.1).mean()
```

```python
plt.scatter(df['date'], df['meantemp'], color='green')
plt.plot(df['date'], xl, color='black', label='Alpha=0.5')
plt.plot(df['date'], xl2, color='red', label='Alpha=0.9')
plt.plot(df['date'], xl3, label='Alpha=0.1')
plt.legend()
plt.show()
```
