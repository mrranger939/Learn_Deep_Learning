```python
import pandas as pd
```

```python
from keras.preprocessing import image
from tensorflow.keras.preprocessing.image import ImageDataGenerator
```

```python
import matplotlib.pyplot as plt
```

```python
img = image.load_img('./train/dog1.jpg', target_size=(300,200))
plt.imshow(img)
```

```python
datagen = ImageDataGenerator(
    rotation_range=30, # rotation
    shear_range=0.2, # distortion
    zoom_range=0.2,
    horizontal_flip=True,
    width_shift_range=0.2,
    height_shift_range=0.2,
    # fill_mode: nearest, constant, reflect : to fill in shifted pixels 
)
```

```python
img = image.img_to_array(img)
```

```python
img.shape
```

```python
input_batch = img.reshape(1,300,200,3)
```

```python
i = 0
for output in datagen.flow(input_batch, batch_size=1, save_to_dir='aug'):
    i += 1
    if i == 10:
        break
    
```
